Pathogenicity scores using the method from [Brandes2023](https://doi.org/10.1038/s41588-023-01465-0):
- `esm_pathogenicity_score` ESM1b LLR score; more negative means more pathogenic (see [Fig2a](https://www.nature.com/articles/s41588-023-01465-0/figures/2) from above)
- `esm_pathogenicity_label` benign/pathogenic labels based on the ESM1b LLR score, thresholded at -7.5 as in the paper

```python
variants = [
  # Known loss-of-function variants from https://www.uniprot.org/uniprotkb/P0A9Q5/entry#phenotypes_variants
  'P0A9Q5/C27A', # Vmax decreases 9-fold
  'P0A9Q5/C30A', # Vmax decreases 140-fold. Loss of nucleic acid-binding
  'P0A9Q5/C46A', # Vmax decreases 11-fold
  'P0A9Q5/C49A', # Vmax decreases 8-fold
  # Likely benign variants eyeballed from the C-terminal domain
  'P0A9Q5/L280A',
  'P0A9Q5/E290A',
  'P0A9Q5/Q300A',
]
mutfunc.query_missense(variants)
```
